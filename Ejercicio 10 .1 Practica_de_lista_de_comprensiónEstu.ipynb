```python
# CONVERTIR -LISTA DE COMPRESION ANIDADAS
# crea una lista points que contiene (en forma de tuplas de dos elementos) la posici√≥n de todos los puntos bidimencionales entre las coordenadas (0, 0) y (5, 10).

points = []
for x in range(0, 5 + 1):
    for y in range(0, 10 + 1):
        points.append((x, y))
print(points)
```
